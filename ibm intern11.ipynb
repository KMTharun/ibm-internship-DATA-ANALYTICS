# Which region & place generates the most sales?

```python
df_places = data[['Country','City','State','Region']]
df_places.head()
```

```python
df_places = data[['City','State','Region','Sales','Profit']]
df_places.head()
```

```python
# Group the data by Region and City and calculate the total sales for each group
region_sales = df_places.groupby(['Region'], as_index=False).sum()
region_sales.sort_values(by='Sales', ascending=False, inplace=True)

# Plot the total sales geProfitnerated by each region and city
plt.figure(figsize=(5,5))
plt.bar(region_sales['Region'], region_sales['Sales'], align='center',)
plt.xlabel("Region")
plt.ylabel("Sales")
plt.title("Sales Generated by Region")
plt.xticks(rotation=50)
plt.show()
region_sales
```

```python
# Group the data by Region and City and calculate the total sales for each group
region_profit = df_places.groupby(['Region'], as_index=False).sum()
region_profit.sort_values(by='Profit', ascending=False, inplace=True)

# Plot the total sales generated by each region and city
plt.figure(figsize=(5,5))
plt.bar(region_profit['Region'], region_profit['Profit'], align='center',)
plt.xlabel("Region")
plt.ylabel("Profit")
plt.title("Profit Generated by Region")
plt.xticks(rotation=90)
plt.show()
region_profit
```

# building model


```python
from sklearn.linear_model import LinearRegression
col=['Ship Mode','Segment','Country','City','State','Postal Code','Region','Category','Sub-Category','Quantity','Discount']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df=data.drop(columns=col,axis=1)
df1=df.copy()
x=df1.drop(["Profit"],axis="columns",inplace=False)
```

```python
y=df1["Profit"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=3/5,random_state=0)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
reg=LinearRegression()
reg=reg.fit(x,y)
```

```python
prediction_lin1=reg.predict(x)
#computing MSE and RMSE
import numpy as np
from sklearn.metrics import mean_squared_error
lin_mse1=mean_squared_error(y,prediction_lin1)
print(np.log(lin_mse1))
```

```python
#R_squared value
r2_lin_test1=reg.score(x_test,y_test)
r2_lin_train1=reg.score(x_train,y_train)
print(r2_lin_test1,r2_lin_train1)
```
