```python
import tkinter as tk
from sklearn import metrics
import matplotlib.pyplot as plt
from tkinter import Tk, filedialog
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor, ExtraTreesRegressor, BaggingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import (
    RandomForestClassifier,
    AdaBoostClassifier,
    GradientBoostingClassifier,
    ExtraTreesClassifier,
    BaggingClassifier,
)
from sklearn.metrics import classification_report, confusion_matrix

# Function to browse and select the dataset file
def browse_file():
    file_path = filedialog.askopenfilename()
    data_entry.delete(0, tk.END)
    data_entry.insert(tk.END, file_path)

# Function to execute the code
def execute_code():
    file_path = data_entry.get()
    # Read the dataset file
    data = pd.read_csv(file_path)

    # Find categorical features
    list_1 = list(data.columns)
    list_cate = []
    for i in list_1:
        if data[i].dtype == "object":
            list_cate.append(i)
    from sklearn.preprocessing import LabelEncoder

    le = LabelEncoder()
    for i in list_cate:
        data[i] = le.fit_transform(data[i])

    # Perform data analysis and modeling operations
    y=data['Ship Mode']
    x=data.drop('Ship Mode',axis=1)

    x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.2)
    # Choose the model and fit the data
    from sklearn.neighbors import KNeighborsClassifier
    knn=KNeighborsClassifier(n_neighbors=7)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
    from sklearn.metrics import r2_score
    from sklearn.metrics import mean_squared_error
    #prediction
    ClassificationReport=classification_report(y_test,y_pred)
    TrainingScore=(knn.score(x_train,y_train)*100)
    accuracy=metrics.accuracy_score(y_test,y_pred)
    
    #SVC
    from sklearn.svm import SVC
    svc = SVC()
    svc.fit(x_train, y_train)
    y_pred=svc.predict(x_test)
    #prediction
    ClassificationReportforSVC=classification_report(y_test,y_pred)
    TrainingScoreforSVC=(svc.score(x_train,y_train)*100)
    accuracyforSVC=metrics.accuracy_score(y_test,y_pred)
    
    #naive bayes
    from sklearn.naive_bayes import GaussianNB
    gnb = GaussianNB()
    gnb.fit(x_train,y_train)
    y_pred=gnb.predict(x_test)
    #prediction
    ClassificationReportforNAIVEBAYES=classification_report(y_test,y_pred)
    TrainingScoreforNAIVEBAYES=(gnb.score(x_train,y_train)*100)
    accuracyforNAIVEBAYES=metrics.accuracy_score(y_test,y_pred)
    
    
    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    rfc=RandomForestClassifier()
    rfc.fit(x_train,y_train)
    y_pred=rfc.predict(x_test)
    #prediction
    ClassificationReportforRandomForestClassifier=classification_report(y_test,y_pred)
    TrainingScoreforRandomForestClassifier=(rfc.score(x_train,y_train)*100)
    accuracyforRandomForestClassifier=metrics.accuracy_score(y_test,y_pred)
    
    #AdaBoostClassifier
    from sklearn.ensemble import AdaBoostClassifier
    adb = AdaBoostClassifier(base_estimator = None)
    adb.fit(x_train,y_train)
    y_pred=adb.predict(x_test)
    #prediction
    ClassificationReportforAdaBoostClassifier=classification_report(y_test,y_pred)
    TrainingScoreforAdaBoostClassifier=(adb.score(x_train,y_train)*100)
    accuracyforAdaBoostClassifier=metrics.accuracy_score(y_test,y_pred)
    
    #Gradient Boosting Classifier
    from sklearn.ensemble import GradientBoostingClassifier
    gbc=GradientBoostingClassifier()
    gbc.fit(x_train,y_train)
    y_pred=gbc.predict(x_test)
    #prediction
    ClassificationReportforGradientBoostingClassifier=classification_report(y_test,y_pred)
    TrainingScoreforGradientBoostingClassifier=(gbc.score(x_train,y_train)*100)
    accuracyforGradientBoostingClassifier=metrics.accuracy_score(y_test,y_pred)
    
    #ExtraTreesClassifier
    from sklearn.ensemble import ExtraTreesClassifier
    etc = ExtraTreesClassifier(n_estimators=100, random_state=0)
    etc.fit(x_train,y_train)
    y_pred=etc.predict(x_test)  
    #prediction
    ClassificationReportforExtraTreesClassifier=classification_report(y_test,y_pred)
    TrainingScoreforExtraTreesClassifier=(etc.score(x_train,y_train)*100)
    accuracyforExtraTreesClassifier=metrics.accuracy_score(y_test,y_pred)
    
    #Bagging Classifier
    from sklearn.ensemble import BaggingClassifier
    from sklearn import tree
    model = BaggingClassifier(tree.DecisionTreeClassifier(random_state=1))
    model.fit(x_train, y_train)
    model.score(x_test,y_test)
    #PREDICTION
    RESULTS = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
    
    
    # Display the results in the result_text widget for knn
    result_text.insert(tk.END, f"KNN \n")
    result_text.insert(tk.END, f"Classification Report is: {ClassificationReport}\n")
    result_text.insert(tk.END, f"Training Score: {TrainingScore}\n")
    result_text.insert(tk.END, f"accuracy: {accuracy}\n")
    
        # Display the results in the result_text widget for SVC
    result_text.insert(tk.END, f"SVC \n")
    result_text.insert(tk.END, f"Classification Report for svc: {ClassificationReportforSVC}\n")
    result_text.insert(tk.END, f"Training Score for svc : {TrainingScoreforSVC}\n")
    result_text.insert(tk.END, f"accuracy for svc: {accuracyforSVC}\n")
    
        # Display the results in the result_text widget for NAIVEBAYES
    result_text.insert(tk.END, f"NAIVEBAYES\n")
    result_text.insert(tk.END, f"Classification Report for NAIVEBAYES is: {ClassificationReportforNAIVEBAYES}\n")
    result_text.insert(tk.END, f"Training Score for NAIVEBAYES: {TrainingScoreforNAIVEBAYES}\n")
    result_text.insert(tk.END, f"accuracy for NAIVEBAYES: {accuracyforNAIVEBAYES}\n")

        # Display the results in the result_text widget for RandomForestClassifier
    result_text.insert(tk.END, f"RandomForestClassifier\n")
    result_text.insert(tk.END, f"Classification Report forRandomForestClassifier: {ClassificationReportforRandomForestClassifier}\n")
    result_text.insert(tk.END, f"Training Score forRandomForestClassifier: {TrainingScoreforRandomForestClassifier}\n")
    result_text.insert(tk.END, f"accuracy forRandomForestClassifier: {accuracyforRandomForestClassifier}\n")
    
        # Display the results in the result_text widget forAdaBoostClassifier
    result_text.insert(tk.END, f"AdaBoostClassifier\n")
    result_text.insert(tk.END, f"Classification Report forAdaBoostClassifier: {ClassificationReportforAdaBoostClassifier}\n")
    result_text.insert(tk.END, f"Training Score forAdaBoostClassifier: {TrainingScoreforAdaBoostClassifier}\n")
    result_text.insert(tk.END, f"accuracy forAdaBoostClassifier: {accuracyforAdaBoostClassifier}\n")
    
        # Display the results in the result_text widget forExtraTreesClassifier
    result_text.insert(tk.END, f"ExtraTreesClassifier\n")
    result_text.insert(tk.END, f"Classification Report forExtraTreesClassifier: {ClassificationReportforExtraTreesClassifier}\n")
    result_text.insert(tk.END, f"Training Score forExtraTreesClassifier: {TrainingScoreforExtraTreesClassifier}\n")
    result_text.insert(tk.END, f"accuracy forExtraTreesClassifier: {accuracyforExtraTreesClassifier}\n")
    
        # Display the results in the result_text widget forGradientBoostingClassifier
    result_text.insert(tk.END, f"GradientBoostingClassifier\n")
    result_text.insert(tk.END, f"Classification Report forGradientBoostingClassifier: {ClassificationReportforGradientBoostingClassifier}\n")
    result_text.insert(tk.END, f"Training Score forGradientBoostingClassifier: {TrainingScoreforGradientBoostingClassifier}\n")
    result_text.insert(tk.END, f"accuracy forGradientBoostingClassifier: {accuracyforGradientBoostingClassifier}\n")
    
        # Display the results in the result_text widget for knn
    result_text.insert(tk.END, f"BaggingClassifier\n")
    result_text.insert(tk.END, f"RESULTS FOR BaggingClassifier : {RESULTS}\n")
    
    
        # Create a figure and axes
    fig, ax = plt.subplots(figsize=(8, 6))
    fig, ax = plt.subplots(figsize=(8, 6))

    # Plot the results
    models = ['SVC', 'Naive Bayes', 'Random Forest', 'AdaBoost', 'Gradient Boosting', 'Extra Trees']
    training_scores = [59.73733583489681,59.19949968730457 , 99.84990619136961, 59.799874921826145,62.11382113821138, 99.84990619136961]
    accuracy_scores = [0.5962981490745373, 0.5882941470735368, 0.5797898949474737, 0.5962981490745373,0.552776388194097 ,0.6003001500750376]

    ax.bar(models, training_scores, label='Training Score')
    ax.bar(models, accuracy_scores, label='Accuracy')

    ax.set_xlabel('Models')
    ax.set_ylabel('Score')
    ax.set_title('Training Score and Accuracy for Different Models')
    ax.legend()
    
    # Create a new window and display the plot
    root = tk.Tk()
    root.title("Plot")
    plot_canvas = FigureCanvasTkAgg(fig, master=root)
    plot_canvas.draw()
    plot_canvas.get_tk_widget().pack()

    # Start the Tkinter event loop
    root.mainloop()


# Create the main window
window = tk.Tk()
window.title("Data Analysis and Modeling")
window.geometry("800x800")

# Create and place the widgets
data_label = tk.Label(window, text="Select Data File:")
data_label.pack()

data_entry = tk.Entry(window, width=400)
data_entry.pack()

browse_button = tk.Button(window, text="provide data set", command=browse_file)
browse_button.pack()

execute_button = tk.Button(window, text="prediction", command=execute_code)
execute_button.pack()

result_text = tk.Text(window, width=800, height=600)
result_text.pack()

# Start the main loop
window.mainloop()
```
